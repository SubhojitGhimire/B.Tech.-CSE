IMPORTING LIBRARIES



```python
import matplotlib.pyplot as plt
import networkx as nx
```

ERDOS-RENYI GRAPH 

Erdos Renyi model is used to create random
networks or graphs on social networking.

n= number of nodes

p = probability of edge creation

```python
G = nx.erdos_renyi_graph(n = 200, p = 0.02)
plt.figure(figsize=(20,20))
nx.draw_networkx(G, with_labels=True)
```

ALBERT-BARABASI-MODEL

A graph of n nodes is grown by attaching new nodes each with m edges that are preferentially attatched to the existing nodes with high degree.

n=number of nodes

m=no of edges to attach from anew node to existing nodes


```python
G = nx.barabasi_albert_graph(n = 1500, m = 5)
plt.figure(figsize=(20,20))
nx.draw_networkx(G, with_labels=True)
```

LancichinettiFortunatoRadicchi benchmark

They have a priori known communities and are used to compare different community detection
methods.

The advantage of the benchmark over other methods is that it accounts for the heterogeneity in
the distributions of node degrees and of community sizes.



```python
from networkx.generators.community import LFR_benchmark_graph
n = 250
tau1 = 3
tau2 = 1.5
mu = 0.1
G = LFR_benchmark_graph(n, tau1, tau2, mu,average_degree=None, min_community = None, seed = None)
plt.figure(figsize=(20,20))
nx.draw_networkx(G, with_labels=True)
```
